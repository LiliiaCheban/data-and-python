Exercise 4 - plot line of best fit for CSIRO
Use the linregress function from scipi.stats to get the slope and y-intercept of the line of best fit.
Plot the line of best fit over the top of the scatter plot.
Make the line go through the year 2050 to predict the sea level rise in 2050.
(Hint: use numpy to create an array from the first year to 2050 with an increment of 1)

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np


url = "https://raw.githubusercontent.com/freeCodeCamp/boilerplate-sea-level-predictor/master/epa-sea-level.csv"
sea_levels = pd.read_csv(url)


# Get the slope and y-intercept of the line of best fit for CSIRO Adjusted Sea Level
regression = linregress(cleaned_sea_levels['Year'], cleaned_sea_levels['CSIRO Adjusted Sea Level'])
slope, intercept = regression.slope, regression.intercept

# Create an extended range for x to predict sea level in 2050
extended_years = np.arange(cleaned_sea_levels['Year'].min(), 2051, 1)

# Predict sea level in 2050 using the line of best fit
predicted_sea_level_2050 = slope * extended_years + intercept

# Plot the scatter graph with line of best fit and prediction for 2050
plt.scatter(cleaned_sea_levels['Year'], cleaned_sea_levels['CSIRO Adjusted Sea Level'], label='CSIRO Adjusted Sea Level')
plt.plot(extended_years, predicted_sea_level_2050, color='red', label='Line of Best Fit (2050 Prediction)')
plt.title('Line of Best Fit for CSIRO Adjusted Sea Level with Prediction for 2050')
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level')
plt.legend()
plt.grid(True)
plt.show()
```

Exercise 5 - plot line of best fit for NOAA
Use the linregress function from scipi.stats to get the slope and y-intercept of the line of best fit.
Plot the line of best fit over the top of the scatter plot.
Make the line go through the year 2030 to predict the sea level rise in 2030
(Hint: use numpy to create an array from the first year to 2030 with an increment of 1 AND remember to use noaa, which has NaN rows removed)

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np


url = "https://raw.githubusercontent.com/freeCodeCamp/boilerplate-sea-level-predictor/master/epa-sea-level.csv"
sea_levels = pd.read_csv(url)

# Clean the data by dropping rows with null values in the 'NOAA' column
cleaned_sea_levels_noaa = sea_levels.dropna(subset=['NOAA Adjusted Sea Level'])

# Get the slope and y-intercept of the line of best fit for NOAA Adjusted Sea Level
regression_noaa = linregress(cleaned_sea_levels_noaa['Year'], cleaned_sea_levels_noaa['NOAA Adjusted Sea Level'])
slope_noaa, intercept_noaa = regression_noaa.slope, regression_noaa.intercept

# Create an extended range for x to predict sea level in 2030
extended_years_noaa = np.arange(cleaned_sea_levels_noaa['Year'].min(), 2031, 1)

# Predict sea level in 2030 using the line of best fit
predicted_sea_level_2030_noaa = slope_noaa * extended_years_noaa + intercept_noaa

# Plot the scatter graph with line of best fit and prediction for 2030
plt.scatter(cleaned_sea_levels_noaa['Year'], cleaned_sea_levels_noaa['NOAA Adjusted Sea Level'], label='NOAA Adjusted Sea Level')
plt.plot(extended_years_noaa, predicted_sea_level_2030_noaa, color='red', label='Line of Best Fit (2030 Prediction)')
plt.title('Line of Best Fit for NOAA Adjusted Sea Level with Prediction for 2030')
plt.xlabel('Year')
plt.ylabel('NOAA Adjusted Sea Level')
plt.legend()
plt.grid(True)
plt.show()
```

Exercise 6 - shorter range (starting 2000)
Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset and using the CSIRO adjusted levels. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np

# Load the sea levels data
url = "https://raw.githubusercontent.com/freeCodeCamp/boilerplate-sea-level-predictor/master/epa-sea-level.csv"
sea_levels = pd.read_csv(url)

# Extract data for the years 2000 through the most recent year
recent_data = sea_levels[sea_levels['Year'] >= 2000]

# Get slope and y-intercept using linregress
regression = linregress(recent_data['Year'], recent_data['CSIRO Adjusted Sea Level'])

# Create an extended range for x to predict sea level in 2050
extended_range = np.arange(2000, 2051, 1)

# Calculate y values for the extended range
predicted_values = regression.slope * extended_range + regression.intercept

# Plot the scatter plot with line of best fit
plt.scatter(sea_levels['Year'], sea_levels['CSIRO Adjusted Sea Level'], label='Actual Data')
plt.plot(extended_range, predicted_values, label='Line of Best Fit', color='red')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level')
plt.title('Sea Level Rise Prediction')

# Show the plot
plt.legend()
plt.show()
```
