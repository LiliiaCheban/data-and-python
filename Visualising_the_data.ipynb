Exercise 4 - Line plot of net migration
Creating a line plot of mean net migration over the years 2015 to 2019

create a new dataframe containing only the five columns holding net migration
create a new data variable to hold the means of the five columns
create a labels variable to hold the keys (column headings)
use plt.plot(x, y) to plot your line chart
*Presentation tips:*
Plot the chart, then add formatting, then show the plot

Rotate the x-axis labels in the plot:
plt.xticks(rotation=45)

Show the grid:
plt.grid()

When all formatting has been done, show the plot:
plt.show()

```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
df = pd.read_excel(url, sheet_name='Country Migration')

# Create a new DataFrame with the five columns containing net migration
migration_columns = ['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']
migration_df = df[migration_columns]

# Create a data variable for the means of the five columns
mean_migration = migration_df.mean()

# Create a labels variable for the keys (years)
labels = migration_columns

# Plot the line chart
plt.plot(labels, mean_migration)

# Add formatting
plt.xticks(rotation=45)
plt.grid()

# Show the plot
plt.show()
```

Exercise 5 - Creating a Bar chart for yearly migration

Create a bar chart which shows the total net migration for the years 2015-2019

split the migration columns into a new dataframe
create a data variable from the max() of the five columns
create a labels variable, this time just create a list of the 5 years ['2015','2016','2017','2018','2019']
plot the bar chart
*Presentation tips*:

use plt.xlabel('Year') and plt.ylabel('Total net migration') to name your axis
display your bar graph with plt.show

```python
import pandas as pd
import matplotlib.pyplot as plt


url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
df = pd.read_excel(url, sheet_name='Country Migration')

# Split the migration columns into a new DataFrame
migration_columns = ['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']
migration_df = df[migration_columns]

# Create a data variable from the sum of the five columns
total_migration = migration_df.sum()

# Create a labels variable with a list of the 5 years
labels = ['2015', '2016', '2017', '2018', '2019']

# Plot the bar chart
plt.bar(labels, total_migration)

# Add axis labels
plt.xlabel('Year')
plt.ylabel('Total net migration')

# Display the bar graph
plt.show()
```

Exercise 6 - creating a bar graph using grouped data

Create a horizontal bar chart of 2019 mean immigration for each income level ('target_country_wb_income')

create a data variable which contains the means, grouped by 'target_country_wb_income'
extract your labels (x) using the .keys() function
use the plt.barh(x, y) to create a horizontal bar graph
add labels to the axes ('Year', 'Net Migration')
show the plot
Try using plt.bar(x,y) - can you see why horizontally is more appropriate?

```python
import pandas as pd
import matplotlib.pyplot as plt


url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
df = pd.read_excel(url, sheet_name='Country Migration')

# Create a data variable containing the means, grouped by 'target_country_wb_income'
data_means = df.groupby('target_country_wb_income')['net_per_10K_2019'].mean()

# Extract labels using the .keys() function
labels = data_means.keys()

# Plot the horizontal bar chart using plt.barh(x, y)
plt.barh(labels, data_means)

# Add labels to the axes
plt.xlabel('Net Migration')
plt.ylabel('Income Level')

# Show the plot
plt.show()
```

Exercise 7 - Creating a pie chart of stress data

Create a pie chart which shows how stressed students who speak different languages are.

To do this:

similar to Exercise 6 - create a variable which groups the means of data by firstlanguage
store the means for 'Stress' in a variable called data
extract your labels using the keys() function
plot your pie chart using plt.pie(data, labels = labels)
write a comment noting anything interesting about the visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt


url = 'https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true'
psych_data = pd.read_csv(url)

# Group the data by firstlanguage and calculate the mean stress for each language
grouped_data = psych_data.groupby('firstlanguage')['Stress'].mean()

# Store the means for 'Stress' in a variable called data
data = grouped_data.values

# Extract labels using the .keys() function
labels = grouped_data.keys()

# Plot the pie chart using plt.pie()
plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Mean Stress Levels by First Language')

# Show the plot
plt.show()
```

Exercise 10 - Create a scatterplot of Wellbeing and Stress with line of best fit
Earlier in exercise 3, we found that Stress significantly predicted Wellbeing.

Create a scatterplot of Wellbeing and Stress data.

create x from the Stress column
create y from the Wellbeing column
use plt.scatter(x,y) to create a scatterplot
add x axis and y axis labels using plt.xlabel('Stress') and plt.ylabel('Wellbeing')
Adding a line of best fit:

create a regression variable and assign the result of using linregress
store the slope and intercept (slope = regression.slope, intercept = regression.intercept)
plt.plot(x, slope*x + intercept)
Show the plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress


url = 'https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true'
psych_data = pd.read_csv(url)

# Extract the 'Stress' and 'Wellbeing' columns
x = psych_data['Stress']
y = psych_data['Wellbeing']

# Create a scatterplot
plt.scatter(x, y)

# Add labels for x and y axes
plt.xlabel('Stress')
plt.ylabel('Wellbeing')

# Adding a line of best fit
regression = linregress(x, y)
slope = regression.slope
intercept = regression.intercept
plt.plot(x, slope * x + intercept, color='red')

# Show the plots
plt.show()
```
